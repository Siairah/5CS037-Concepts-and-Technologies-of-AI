 Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual
ization to show how the survival distributions vary by gender.

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('/content/drive/My Drive/WEEK2AI/titanic.csv')

#Calculate mean survival rates for different groups in the 'Sex' column
mean_survival_rates = dataset.groupby('Sex')['Survived'].mean()
print(mean_survival_rates)
#Draw the visualization
plt.figure(figsize=(10, 6))
dataset.groupby('Sex')['Survived'].mean().plot(kind='bar', color=['blue', 'orange'])
plt.title('Survival Rates by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
[]
```

 Draw a visualization that breaks your visualization from Exercise 3 down by the port of embarkation (’Em
barked’). In this instance, compare the ports ’C’ (Cherbourg), ’Q’ (Queenstown), and ’S’ (Southampton)

```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('/content/drive/My Drive/WEEK2AI/titanic.csv')

embarked_dummies = pd.get_dummies(dataset['Embarked'], prefix='Embarked')

#Add the new columns to the original DataFrame
dataset = pd.concat([dataset, embarked_dummies], axis=1)

# Drop the original 'Embarked' column
dataset.drop('Embarked', axis=1, inplace=True)

# Group by 'Embarked' columns and calculate the mean survival rate
embarked_survival = dataset.groupby(['Embarked_C', 'Embarked_Q', 'Embarked_S'])['Survived'].mean()

# Plot the data
embarked_survival.plot(kind='bar', figsize=(10, 6))
plt.title('Survival Rate by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Survival Rate')
plt.xticks(ticks=[0, 1, 2], labels=['Cherbourg (C)', 'Queenstown (Q)', 'Southampton (S)'])
plt.show()
```
