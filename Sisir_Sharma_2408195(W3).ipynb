1. Create a new column, age
 group, that categorizes passengers into "child" (age < 18), "adult" (age
 18{64), and "senior" (age 65 and above).

```python
import pandas as pd

df = pd.read_csv("/content/drive/My Drive/WEEK2AI/titanic.csv")

df['Age'] = df['Age'].fillna(df['Age'].median())
age_group_labels = ['Child', 'Adult', 'Senior']
bins = [-1, 18, 65, float('inf')]

df['age_group'] = pd.cut(df['Age'], bins=bins, labels=age_group_labels, right=False)
print(df[['Age', 'age_group']].head())

```

2. Calculate the total number of passengers on the Titanic.

```python
total_passengers = len(df)

print(f"Total number of passengers on the Titanic: {total_passengers}")
```

 3. Count the number of passengers in each age group.

```python
# Count the number of passengers in each age group
age_group_counts = df['age_group'].value_counts()

print(f"Number of passengers in each age group:\n{age_group_counts}")
```

4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
 in each age group

```python

total_passengers = len(df)
age_group_counts = df['age_group'].value_counts()
age_group_proportions = age_group_counts / total_passengers

print(f"Proportion of passengers in each age group:\n{age_group_proportions}")
```

5. Display the proportion as a percentage

```python

age_group_proportions = age_group_counts / total_passengers

age_group_percentage = age_group_proportions * 100

print(f"Proportion of passengers in each age group (as a percentage):\n{age_group_percentage}")
```
