Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
# Calculate IQR for Happiness Score
score_q1 = south_asia_df['score'].quantile(0.25)
score_q3 = south_asia_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1
score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

# Calculate IQR for GDP per Capita
gdp_q1 = south_asia_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = south_asia_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1
gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# Outliers for Score and GDP per Capita
score_outliers = south_asia_df[(south_asia_df['score'] < score_lower_bound) | (south_asia_df['score'] > score_upper_bound)]
gdp_outliers = south_asia_df[(south_asia_df['Log GDP per capita'] < gdp_lower_bound) | (south_asia_df['Log GDP per capita'] > gdp_upper_bound)]

# Combine outliers
outliers = pd.concat([score_outliers, gdp_outliers]).drop_duplicates()

print("Outliers based on Happiness Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

2. Define outliers using the 1.5 × IQR rule.

Ans:
Calculate the Quartiles (Q1 and Q3).The value below which 25% of the data points fall. It is the median of the lower half of the data.
Q3 (Third Quartile): The value below which 75% of the data points fall. It is the median of the upper half of the data.
Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1
The IQR represents the range within which the middle 50% of the data lies.Determine the Lower Bound
Lower Bound = Q1 - 1.5 × IQR
This is the threshold below which any data point is considered an outlier on the lower end.Determine the Upper Bound
Upper Bound = Q3 + 1.5 × IQR
This is the threshold above which any data point is considered an outlier on the higher end.
Identify Outliers
Any data point that falls below the lower bound or above the upper bound is considered an outlier.

6. Outlier Detection:
 • Identify outlier countries in both regions based on Score and GDP per Capita.

```python
# Calculate Q1 and Q3
Q1_score = south_asia_df['score'].quantile(0.25)
Q3_score = south_asia_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asia_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asia_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Determine the lower and upper bounds for outliers
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers
south_asia_score_outliers = south_asia_df[(south_asia_df['score'] < lower_bound_score) | (south_asia_df['score'] > upper_bound_score)]
south_asia_gdp_outliers = south_asia_df[(south_asia_df['Log GDP per capita'] < lower_bound_gdp) | (south_asia_df['Log GDP per capita'] > upper_bound_gdp)]
print(f"As per GDP outlier countries in South Asia:\n {south_asia_gdp_outliers[['Country name', 'Log GDP per capita']]}")
print(f"As per score outlier countries in South Asia:\n {south_asia_score_outliers[['Country name', 'score']]}")



```
